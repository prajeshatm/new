```python
#import dataset basketball_awards_coaches.csv
import pandas as pd
df=pd.read_csv('basketball_awards_coaches.csv')
df
```

```python
#display the number of rows and columns in the dataset
df.shape
```

```python
#NaN means there is no tie for awrd in that year.So NaN is replaced by 'no tie'
df[['note']] = df[['note']].fillna(value='no tie')
```

```python
#display first ten rows
df.head(10)
```

```python
# importing matplotlib module 
from matplotlib import pyplot as plt
# Y-axis values
y = df['lgID']
# Function to plot histogram
plt.hist(y) # shows which award obtained the number of times each category is awarded in the given years
# Function to show the plot
plt.ylabel("Number of times awarded")
plt.xlabel("Awards")
plt.show()
```

```python
#Find which coach got both the award most of the time
df.groupby('coachID')['award'].count()
```

```python
#Shows the number of persons received award in each year
df.groupby('year')['award'].count()
```

```python
# importing matplotlib module 
from matplotlib import pyplot as plt
# Y-axis values
y = df['year']
# Function to plot histogram
plt.hist(y) 
# Function to show the plot
plt.ylabel("Number of persons awarded")
plt.xlabel("Year")
plt.show()# Shows National and American basketball associations gave award to more number of persons during the period of 1965-1975
```
